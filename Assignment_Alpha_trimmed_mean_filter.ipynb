```python
import cv2
import numpy as np
from matplotlib import pyplot as pit
from google.colab.patches import cv2_imshow
#from google.colab import drive
#drive.mount('/content/drive')
```

```python
image = cv2.imread('/content/image091.jpg',0)
print(image)
print(image.shape)
cv2_imshow(image)
height,width=image.shape
```

```python
padding_value = 1
padding_image = np.zeros((height + 2 * padding_value, width + 2 * padding_value), dtype=np.uint8)
padding_image[padding_value:-padding_value, padding_value:-padding_value] = image
cv2_imshow(padding_image)
```

```python
filter_size = 3
d_value = 1
padding = filter_size // 2
filtered1_image = np.zeros_like(image)
for i in range(padding, image.shape[0] + padding):
    for j in range(padding, image.shape[1] + padding):
        neighborhood = padding_image[i - padding:i + padding + 1, j - padding:j + padding + 1]
        sorted_vals = np.sort(neighborhood.flatten())
        trimmed_vals = sorted_vals[d_value:-d_value] 
        mean = np.mean(trimmed_vals)
        filtered1_image[i - padding, j - padding] = mean

        

filter_size = 3
d_value = 2
padding = filter_size // 2
filtered2_image = np.zeros_like(image)
for i in range(padding, image.shape[0] + padding):
    for j in range(padding, image.shape[1] + padding):
        neighborhood = padding_image[i - padding:i + padding + 1, j - padding:j + padding + 1]
        sorted_vals = np.sort(neighborhood.flatten())
        trimmed_vals = sorted_vals[d_value:-d_value] 
        mean = np.mean(trimmed_vals)
        filtered2_image[i - padding, j - padding] = mean




filter_size = 3
d_value = 4
padding = filter_size // 2
filtered3_image = np.zeros_like(image)
for i in range(padding, image.shape[0] + padding):
    for j in range(padding, image.shape[1] + padding):
        neighborhood = padding_image[i - padding:i + padding + 1, j - padding:j + padding + 1]
        sorted_vals = np.sort(neighborhood.flatten())
        trimmed_vals = sorted_vals[d_value:-d_value] 
        mean = np.mean(trimmed_vals)
        filtered3_image[i - padding, j - padding] = mean



print("different images with different d value:")    
print("")
print("d = 1")   
cv2_imshow(filtered1_image)
print("")
print("d = 2") 
cv2_imshow(filtered2_image)
print("")
print("d = 4") 
cv2_imshow(filtered3_image)





```
