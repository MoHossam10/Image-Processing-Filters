```python
import cv2
import numpy as np
from matplotlib import pyplot as pit
from google.colab.patches import cv2_imshow
#from google.colab import drive
#drive.mount('/content/drive')
```

```python
image = cv2.imread('/content/image091.jpg',0)
print(image)
print(image.shape)
cv2_imshow(image)
height,width=image.shape
```

```python
padding_value = 1
padding_image = np.zeros((height + 2 * padding_value, width + 2 * padding_value), dtype=np.uint8)
padding_image[padding_value:-padding_value, padding_value:-padding_value] = image
cv2_imshow(padding_image)
print(padding_image.shape)
```

```python
filter_size = 3
padding = filter_size // 2
new_image = np.zeros((height, width), np.uint8)
for i in range(padding, height+padding):
    for j in range(padding, width+padding):
      neighborhood = padding_image[i - padding:i + padding + 1, j - padding:j + padding + 1]
      min_value = np.min(neighborhood)
      max_value = np.max(neighborhood)
      average = (min_value + max_value) / 2
      new_image[i - padding, j - padding] = average
cv2_imshow(image)
print("")   
cv2_imshow(new_image)      
```
